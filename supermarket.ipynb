```python
df = pd.read_csv('supermarket_sales.csv')
df.head()
df
males_quan   = df.loc[df['Gender'] == 'Male',   'Quantity']
females_quan = df.loc[df['Gender'] == 'Female', 'Quantity']
a = []
b = []
for i in range(1, 11):
    m = males_quan.loc[males_quan == i]
    m = m.sum() / i
    a.append(m)

    m = females_quan.loc[females_quan == i]
    m = m.sum() / i
    b.append(m)
a

```

```python


fig, ax = plt.subplots(figsize=(12, 6))  # Specify the figure size

ax.bar(datetime_dates, gross, color=colors, width=0.25)
ax.set_xlabel('Dates')
ax.set_ylabel('gross income')
ax.set_title('date + gross income relation')
plt.gca().xaxis.set_major_locator(mdates.MonthLocator())
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m'))
plt.xticks(rotation=45)
plt.subplots_adjust(left=0.1, right=0.9, top=0.9, bottom=0.1)  # Modify the margins as desired
plt.show()
```

```python
def getCount(l_type):
    gender_list = df.loc[df['Product line'] == l_type, 'Gender']
    males = 0
    females = 0
    for g in gender_list:
        if g == 'Male':
            males = males + 1
        else:
            females = females + 1
    print('type' , l_type , "\n")
    print('males count:' , males , '\n')
    print('females count:' , females , '\n')

for t in ss.unique():
    getCount(t)

```
